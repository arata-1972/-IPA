```python
# ファイルを読み込み、内容を処理するPythonコード
file_path = '/Users/arata1972/Downloads/pheng-im-pio.txt'



# 台羅とIPAの対応を示す辞書を作成
tai_lo_ipa_dict = {}

# ファイルを読み込み、各行を処理して辞書に追加
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

for line in lines:
    columns = line.split()
    if len(columns) < 9:
        continue
    tai_lo = ''.join([c for c in columns[0] if not c.isdigit() and c != 'ⁿ'])  # 声調を除く
    ipa = columns[7]
    if tai_lo not in tai_lo_ipa_dict:
        tai_lo_ipa_dict[tai_lo] = [ipa]

# 結果を表示
import json

# カスタムのJSONエンコーダを使用して1行に表示
class CompactJSONEncoder(json.JSONEncoder):
    def iterencode(self, o, _one_shot=False):
        if isinstance(o, (list, tuple)):
            return "[" + ", ".join(map(self.encode, o)) + "]"
        return super().iterencode(o, _one_shot)

print(json.dumps(tai_lo_ipa_dict, ensure_ascii=False, indent=None, separators=(',', ': '), cls=CompactJSONEncoder))


```
