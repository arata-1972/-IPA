```python
import re

# ファイルのパスを設定
file_path = '/Users/arata1972/Downloads/filtered_pheng_im_pio.txt'

# 結果を格納する辞書
conversion_dict = {}

# ファイルを読み込む
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# 正規表現パターンを作成
pattern = re.compile(r'([\w-]+)\s+([^\s]+)')

# 各行を処理
for line in lines:
    match = pattern.match(line)
    if match:
        # 台羅
        tailo = match.group(1).strip()
        # 国際音標（IPA）
        ipa = match.group(2).strip()
        
        # 声調を除去
        tailo_no_tone = re.sub(r'[áàâäãā]', 'a', tailo)
        tailo_no_tone = re.sub(r'[éèêëẽē]', 'e', tailo_no_tone)
        tailo_no_tone = re.sub(r'[íìîïĩī]', 'i', tailo_no_tone)
        tailo_no_tone = re.sub(r'[óòôöõō]', 'o', tailo_no_tone)
        tailo_no_tone = re.sub(r'[úùûüũū]', 'u', tailo_no_tone)
        tailo_no_tone = re.sub(r'[ÁÀÂÄÃĀ]', 'A', tailo_no_tone)
        tailo_no_tone = re.sub(r'[ÉÈÊËẼĒ]', 'E', tailo_no_tone)
        tailo_no_tone = re.sub(r'[ÍÌÎÏĨĪ]', 'I', tailo_no_tone)
        tailo_no_tone = re.sub(r'[ÓÒÔÖÕŌ]', 'O', tailo_no_tone)
        tailo_no_tone = re.sub(r'[ÚÙÛÜŨŪ]', 'U', tailo_no_tone)
        
        # IPAの声調記号を除去
        ipa_no_tone = re.sub(r'[˥˧˨˩]', '', ipa)
        
        # 辞書に追加
        conversion_dict[tailo_no_tone] = [ipa_no_tone]

# 結果を表示
for key, value in conversion_dict.items():
    print(f'"{key}": {value},')
```

```python
import pandas as pd

# ファイルのパス
file_path = '/Users/arata1972/Downloads/pheng-im-pio.txt'

# ファイルを読み込み、必要な行を削除
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# 0, 1, 2, 3, 4, 5, 8行目を削除
lines_to_keep = [line for i, line in enumerate(lines) if i not in [0, 1, 2, 3, 4, 5, 8]]

# データフレームに読み込み
data = pd.DataFrame([line.strip().split('\t') for line in lines_to_keep])

# 不要な列を削除
data = data[[6, 7]]  # 台羅と國際音標の列を保持

# 結果を新しいファイルに保存
output_file_path = '/Users/arata1972/Downloads/filtered_pheng_im_pio.txt'
data.to_csv(output_file_path, sep='\t', header=False, index=False)

print("処理が完了しました。結果は 'filtered_pheng_im_pio.txt' に保存されました。")
```
